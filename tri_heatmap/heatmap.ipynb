```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.tri as tri
from matplotlib.colors import Normalize
%matplotlib inline
```

## Data Preparation
This method may be sensitive to this

```python
zmin = 0
zmax = 1

xlist = np.random.uniform(low=0, high=100, size=(1000,))
ylist = np.random.uniform(low=0, high=10, size=(1000,))
zlist = np.random.uniform(low=zmin, high=zmax, size=(1000,))

xyz = pd.DataFrame({'x' : xlist, 'y' : ylist, 'z' : zlist})

x = xyz['x']
y = xyz['y']
z = xyz['z']

xyz.sample(5)
```

```python
norm = Normalize(vmin=zmin, vmax=zmax, clip=True)

triang = tri.Triangulation(x, y)
```

## Tricontourf
#### Note colorbar is not smoothed using this method

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

surf = ax.tricontourf(triang, z, norm=norm, cmap=matplotlib.cm.viridis)

cbar = fig.colorbar(surf)
cbar.ax.set_ylabel('z label', size=20, labelpad=15)
ax.set_xlabel('x label', fontsize=20, labelpad=15)
ax.set_ylabel('y label', fontsize=20, labelpad=15)
ax.set_title('Title', fontsize=20, pad=15)
plt.show()
```

## Tripcolor
#### Note colorbar IS smoothed using this method
### Flat Shading

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

surf = ax.tripcolor(triang, z, norm=norm, cmap=matplotlib.cm.viridis, shading='flat')

cbar = fig.colorbar(surf)
cbar.ax.set_ylabel('z label', size=20, labelpad=15)
ax.set_xlabel('x label', fontsize=20, labelpad=15)
ax.set_ylabel('y label', fontsize=20, labelpad=15)
ax.set_title('Title', fontsize=20, pad=15)
plt.show()
```

## Tripcolor 
### Gouraud Shading

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111)

surf = ax.tripcolor(triang, z, norm=norm, cmap=matplotlib.cm.viridis, shading='gouraud')

cbar = fig.colorbar(surf)
cbar.ax.set_ylabel('z label', size=20, labelpad=15)
ax.set_xlabel('x label', fontsize=20, labelpad=15)
ax.set_ylabel('y label', fontsize=20, labelpad=15)
ax.set_title('Title', fontsize=20, pad=15)
plt.show()
```
